```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
n = np.random.normal(0,1,100)
plt.hist(n)
plt.show()
```

```python
def gaussian(x, NO, mu, sigma):
    return NO*np.exp(-0.5*((x-mu)/sigma)**2)

N = 1000
test_x = np.random.normal(size=N)
test_sigma = 0.02*np.random.normal(size=len(test_x))
b = np.linspace(-4,4,19)
count,bins,ignored = plt.hist(test_x,bins=b)

mu_fit, sigma_fit = norm.fit(test_x)

renorm = N*(b[1]-b[0])
x = np.linspace(-4., 4., 120)
N0 = 1/(sigma_fit*np.sqrt(2*np.pi))
plt.plot(x,renorm*gaussian(x,N0,mu_fit,sigma_fit))
plt.show()
```
