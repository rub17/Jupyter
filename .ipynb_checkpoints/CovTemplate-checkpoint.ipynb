```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
from scipy.stats import skew
from scipy.stats.stats import pearsonr

%config InlineBackend.figure_format = 'retina'
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12.0,6.0)
```

```python
C1 = pd.read_csv("JERCov_New.csv",header=None)
```

```python
C1 = C1.iloc[:,:6]
```

```python
C2 = pd.read_csv("TemplateCorr.csv",header=None)
```

```python
C2 = C2.iloc[1:,:]
```

```python
C1, C2
```

```python
sigmas = pd.Series([0.09293055,0.0148728,0.0702082,0.07733995,0.0260997,0.0701606]) #TRExFitter Sigmas
S = pd.DataFrame(np.diag(sigmas))
```

```python
C2 = C2.set_index(pd.Index([x for x in range(0,6)]))
```

```python
C2Cov = S.dot(C2.dot(S)) #Now it's a covariance matrix!
```

```python
C2Cov
```

```python
Ctot = C1 + C2Cov
```

```python
POIs = {0:"Px_t",1:"Py_t",2:"Pz_t",3:"Px_tbar",4:"Py_tbar",5:"Pz_tbar"}
Ctot = Ctot.rename(index = POIs, columns = POIs)
```

```python
fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
covPlot = sns.heatmap(Ctot*100, annot=True,fmt=".1f")
figure = covPlot.get_figure()
#figure.savefig('Covariance_Inclusive.pdf')
```

```python
Ctot
```

```python
D = np.diag(np.sqrt(np.diag(Ctot)))
```

```python
Crho = np.linalg.inv(D) @ Ctot @ np.linalg.inv(D) # Getting corrleation matrix
```

```python
Crho = Crho.rename(index = POIs, columns = POIs)
Crho
```

```python
fig=plt.figure(figsize=(14, 10), dpi= 80)
covPlot = sns.heatmap(Crho*100, annot=True,fmt=".1f",cmap="inferno")
sns.set(font_scale=1.5)
figure = covPlot.get_figure()
figure.savefig("TemplateCorr.pdf")
```
