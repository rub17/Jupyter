```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from scipy.stats import skew
from scipy.stats.stats import pearsonr
import seaborn as sns

%config InlineBackend.figure_format = 'retina'
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12.0,6.0)
```

```python
labels = ['polx_1','polx_2','polx_3','polx_4','polx_5','polx_6','polx_7','polx_8',
          'poly_1','poly_2','poly_3','poly_4','poly_5','poly_6','poly_7','poly_8',
          'polz_1','polz_2','polz_3','polz_4']
SysFile = open('systematicsList.txt')
SysNames = SysFile.read().splitlines()
df = pd.read_csv('input/sysData_top.csv',index_col=False,names=labels)
df = pd.DataFrame(data=df.values, index=SysNames,columns=labels)
df.shape
```

```python
df.head()
```

```python
df_up = df[(df.index.str.contains('up') | df.index.str.contains('UP') | df.index.str.contains('Up'))]
df_down = df[df.index.str.contains('down') | df.index.str.contains('DOWN') | df.index.str.contains('Down')]
df_others = df[~df.index.isin(df_up.index | df_down.index)]
#df_up.shape, df_down.shape,df_others.shape
```

```python
sysNamesUP = []
for index,row in df_up.iterrows():
    if ('down' not in index):
        if ('Pileup' not in index):
            a = index.replace('UP',"")
            a = a.replace('up',"")
            a = a.replace('Up',"")
            sysNamesUP.append(a)
        else:
            print(index)
            a = index.replace('1up',"")
            sysNamesUP.append(a)
    else:
        print(index)
        df_up = df_up.drop(index = index)

sysNamesDOWN = []
for index,row in df_down.iterrows():
    if ('up' not in index):
        a = index.replace('DOWN',"")
        a = a.replace('down',"")
        a = a.replace('Down',"")
        sysNamesDOWN.append(a)
    else:
        if ('Pileup' in index):
            sysNamesDOWN.append(index.replace('1down',""))
        else:
            print("Problem with: " + index)
for index,name in enumerate(sysNamesUP):
    if index > len(sysNamesDOWN)-1:
        print (name)
    else:
        if name != sysNamesDOWN[index]:
            print (name,sysNamesDOWN[index])
            
df_up = df_up.set_index([sysNamesUP])
df_down = df_down.set_index([sysNamesDOWN])
df_up.shape, df_down.shape
```

```python
#Taking the last 8 Up variations elsewhere
df_up1side = df_up.tail(8)
df_up1side
```

```python
df_up = df_up.drop(df_up.tail(8).index)
df_up
```

```python
#df_nominal = pd.DataFrame(data=df.loc['Nominal'],column=df.loc['Nominal'].index)\
df_nominal = pd.DataFrame()
df_nominal = df_nominal.append(df.iloc[0],ignore_index=True)
df_nominal
```

```python
df_others[df_others.index.str.contains('PDF')].tail()
```

```python
dictOthers = {}
for i in range (0,4):
    dictOthers["df_others_PDF{x}".format(x=i)] = df_others[df_others.index.str.contains('PDF')][i*31:(i+1)*31]
```

```python
df_rest = df_others[~df_others.index.str.contains('PDF')].drop("Nominal")
df_up.shape,df_down.shape,dictOthers["df_others_PDF0"].shape,dictOthers["df_others_PDF1"].shape,dictOthers["df_others_PDF2"].shape,dictOthers["df_others_PDF3"].shape,df_rest.shape,df_up1side.shape
```

```python
df_VUP = df_up - df_nominal.iloc[0].values
df_VDOWN = df_down - df_nominal.iloc[0].values
df_VPDF = dictOthers["df_others_PDF0"] - df_nominal.iloc[0].values
df_VPDF = df_VPDF.append(dictOthers["df_others_PDF1"] - df_nominal.iloc[0].values)
df_VPDF = df_VPDF.append(dictOthers["df_others_PDF2"] - df_nominal.iloc[0].values)
df_VPDF = df_VPDF.append(dictOthers["df_others_PDF3"] - df_nominal.iloc[0].values)
df_Vrest = df_rest - df_nominal.iloc[0].values
df_Vup1side = df_up1side - df_nominal.iloc[0].values
```

```python
C_UpDown = (df_VUP.values.transpose().dot(df_VUP.values) + df_VDOWN.values.transpose().dot(df_VDOWN.values))/2 #Averaging up and down effects
C_PDF = (df_VPDF.values.transpose().dot(df_VPDF.values))/31 #Averaging the PDF sets?
C_rest = df_Vrest.values.transpose().dot(df_Vrest.values)
C_up1side = df_Vup1side.values.transpose().dot(df_Vup1side.values)
```

```python
C = C_UpDown + C_PDF + C_rest + C_up1side
```

```python
pd.DataFrame(C,index = labels, columns = labels)
```

```python
fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
covPlot = sns.heatmap(C*1e6, annot=True,fmt=".1f")
figure = covPlot.get_figure()
figure.savefig('Covariance_top.pdf')
```
